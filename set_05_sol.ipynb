In the multivariate case, a gaussian distribution is defined via a mean and a covrience matrix. Here the covarience matrix is the equivalant of varience in higher dimensions.
```python
#covariance matrix_sample 1

x1 = sample1['x']
y1 = sample1['y']
cov1 = np.vstack((x1,y1))
print(np.cov(cov1))
```

```python
#correlation matrix_sample 1

np.array(sample1.corr())
```

```python
#covariance matrix_sample 2

x2 = sample1['x']
y2 = sample2['y']
cov2 = np.vstack((x2,y2))
print(np.cov(cov2))
```

```python
#correlation matrix_sample 2

np.array(sample2.corr())
```

```python
#covariance matrix_sample 3

x3 = sample3['x']
y3 = sample3['y']
cov3 = np.vstack((x3,y3))
print(np.cov(cov3))
```

```python
#correlation matrix_sample 3

np.array(sample3.corr())
```
